### Problem 45: Triangular, Pentagonal, and Hexagonal
<p>Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:</p>
<table><tr><td>Triangle</td>
<td> </td>
<td>$T_n=n(n+1)/2$</td>
<td> </td>
<td>$1, 3, 6, 10, 15, \dots$</td>
</tr><tr><td>Pentagonal</td>
<td> </td>
<td>$P_n=n(3n - 1)/2$</td>
<td> </td>
<td>$1, 5, 12, 22, 35, \dots$</td>
</tr><tr><td>Hexagonal</td>
<td> </td>
<td>$H_n=n(2n - 1)$</td>
<td> </td>
<td>$1, 6, 15, 28, 45, \dots$</td>
</tr></table><p>It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.</p>
<p>Find the next triangle number that is also pentagonal and hexagonal.</p>


```python
def triangle(n):
    return n*(n+1)/2

def pentagonal(n):
    return n*(3*n -1)/2

def hexagonal(n):
    return n*(2*n - 1)

def solve_quadratic(a,b,c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        x1 = (-b + discriminant**0.5)/(2*a)
        x2 = (-b - discriminant**0.5)/(2*a)
        return max(x1,x2) % 1 == 0 and int(max(x1,x2))
    elif discriminant == 0:
        x1 = -b/(2*a)
        return x1 % 1 == 0 and int(x1)
    else:
        return False


def is_triangle(k):
    a,b,c = 1, 1, -2*k
    if solve_quadratic(a,b,c):
        return True
    return False

def is_pentagonal(k):
    a,b,c = 3, -1, -2*k
    if solve_quadratic(a,b,c):
        return True
    return False

def is_hexagonal(k):
    a,b,c = 2, -1, -k
    if solve_quadratic(a,b,c):
        return True
    return False
```

```python
def res():
    i = 143
    while True:
        i += 1
        h = hexagonal(i)
        if is_triangle(h) and is_pentagonal(h):
            return h
```
