### Problem 39: Integer Right Triangles
<p>If $p$ is the perimeter of a right angle triangle with integral length sides, $\{a, b, c\}$, there are exactly three solutions for $p = 120$.</p>
<p>$\{20,48,52\}$, $\{24,45,51\}$, $\{30,40,50\}$</p>
<p>For which value of $p \le 1000$, is the number of solutions maximised?</>
>

```python
def triangle_create(n,max_perim=1000):
    for m in range(1,n):
        a = n**2 - m**2
        b = 2*m*n
        c = n**2 + m**2
        k = 1
        while True:
            p = k*a + k*b + k*c
            if p > max_perim:
                break

            l = [k*a, k*b, k*c]
            a_ = min(l)
            l.remove(a_)
            c_ = max(l)
            l.remove(c_)
            b_ = max(l)
            yield [a_, b_, c_]
            k += 1
```

```python
triangle_dict = {}
res = []
seen = []
for i in range(1,1000):
    for x in triangle_create(i,max_perim=1000):
        if x is None: 
            continue
        elif sum(x) not in triangle_dict:
            triangle_dict[sum(x)] = [x]
        elif x not in triangle_dict[sum(x)]:
            triangle_dict[sum(x)] += [x]
```

```python
max_key = max(triangle_dict, key=lambda k: len(triangle_dict[k]))
```
