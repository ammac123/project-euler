```python
import sympy
from decimal import *
getcontext().prec = 100
k = -999
best = 0
best_val = sympy.cosh(sympy.pi * 0.5)
while k < 0:
    
    new = sympy.cos(sympy.pi * sympy.sqrt(k))

    if abs(new.evalf(n=1000) - round(new.evalf(n=1000))) < abs(best_val.evalf(n=1000) - round(best_val.evalf(n=1000))):
        print(k, new.evalf(1000) - round(new.evalf(1000)))
        best = k
        best_val = new
    k += 1

print(best, best_val)
```
